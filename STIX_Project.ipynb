```python
from stix2 import Indicator

# create an Indicator instance 
indicator = Indicator(
    name="Indicator",
    labels=["malicious-activity"],
    description="description",
    pattern="[ipv4-addr:value = '192.0.2.1']",
    pattern_type="stix"
)

# print the indicator object
print(indicator)

from stix2 import Vulnerability

# create a Vulnerability 
vulnerability = Vulnerability(
    name="Vulnerability",
    external_references=[
        {
            "source_name": "CVE",
            "external_id": "CVE-2022-1234",
        }
    ],
)
# print the vulnerability object
print(vulnerability)

from stix2 import IntrusionSet
# create an IntrusionSet instance
intrusion_set = IntrusionSet(
    name="Intrusion Set",
    external_references=[
        {
            "source_name": "Attack",
            "external_id": "IS123",
        }
    ],
)
# print the intrusion set object
print(intrusion_set)

import random
from stix2 import Malware
# prompt the user for input
randomize = input("Do you want to randomize the data of the malware? Enter 'yes' or 'no': ")
if randomize.lower() == "yes":
    # generate random values for the properties
    name = "Malware" + str(random.randint(1, 100))
    description = "This is a description of " + name
    is_family = str(random.choice([True, False]))
else:
    # prompt the user for input
    name = input("Enter the name of the malware: ")
    description = input("Enter a description of the malware: ")
    is_family = input("Is the malware a member of a malware family? Enter 'True' or 'False': ")
# create a Malware instance 
malware = Malware(
    name=name,
    description=description,
    is_family=is_family.lower() == "true",
)
# print malware object
print(malware)


import random
from stix2 import ThreatActor
# prompt the user for input
randomize = input("Do you want to randomize the data for the threat actor? Enter 'yes' or 'no': ")

if randomize.lower() == "yes":
    # generate random values for the properties
    name = "ThreatActor" + str(random.randint(1, 100))
    description = "This is a description of " + name
    sophistication = str(random.randint(1, 5))
else:
    # prompt the user for input
    name = input("Enter the name of the threat actor: ")
    description = input("Enter a description of the threat actor: ")
    sophistication = input("Enter the sophistication level of the threat actor (1-5): ")
# create a ThreatActor instance with the properties
threat_actor = ThreatActor(
    name=name,
    description=description,
    sophistication=int(sophistication),
)
# print the created threat actor object
print(threat_actor)

from stix2 import Identity
from faker import Faker
fake = Faker()
# Prompt the user if they want to randomize the data
response = input("Do you want to randomize the data for the Identity? (yes/no): ")
# Create Identity object
if response.lower() == 'yes':
    identity = Identity(
        name=fake.name(),
        identity_class="organization",
        sectors=[fake.job()],
        contact_information=fake.email()
    )
else:
    # Get user input for Identity properties
    name = input("Enter the name for the Identity: ")
    identity_class = input("Enter the identity class for the Identity (e.g. individual, organization, etc.): ")
    sectors = [input("Enter a sector for the Identity (press enter to skip): ")]
    contact_information = input("Enter the contact information for the Identity: ")

    # Create Identity object
    identity = Identity(
        name=name,
        identity_class=identity_class,
        sectors=sectors,
        contact_information=contact_information
    )
print(identity)

# Defines the properties for the STIX operation
import random
from stix2 import Campaign
# prompt the user for input
randomize = input("Do you want to randomize the data of the campaign? Enter 'yes' or 'no': ")
if randomize.lower() == "yes":
    # generate random values 
    name = "Campaign" + str(random.randint(1, 100))
    objective = "This is the objective of " + name
    first_seen = "2022-01-01T00:00:00Z"
    last_seen = "2022-01-31T23:59:59Z"
else:
    # prompt the user for input
    name = input("Enter the name of the campaign: ")
    objective = input("Enter the objective of the campaign: ")
    first_seen = input("Enter the first seen timestamp (in 2022-01-01T00:00:00Z format): ")
    last_seen = input("Enter the last seen timestamp (in 2022-01-01T00:00:00Z format): ")
# create a Campaign instance 
campaign = Campaign(
    name=name,
    objective=objective,
    first_seen=first_seen,
    last_seen=last_seen,
)
# print the created campaign 
print(campaign)
 
import random
from stix2 import AttackPattern, KillChainPhase
# prompt the user for input
randomize = input("Do you want to randomize the data for the atack pattern? Enter 'yes' or 'no': ")
if randomize.lower() == "yes":
    # generate random values for the properties
    name = "AttackPattern" + str(random.randint(1, 100))
    description = "This is the description of " + name
    kill_chain_phases = [KillChainPhase(phase_name="kill-chain-phase-1", 
                                        kill_chain_name="kill-chain-1")]
else:
    # prompt the user for input
    name = input("Enter the name of the attack pattern: ")
    description = input("Enter the description of the attack pattern: ")
    phase_name = input("Enter the name of the kill chain phase: ")
    kill_chain_name = input("Enter the name of the kill chain: ")
    kill_chain_phases = [KillChainPhase(phase_name=phase_name, kill_chain_name=kill_chain_name)]
# create an AttackPattern instance with the properties
attack_pattern = AttackPattern(
    name=name,
    description=description,
    kill_chain_phases=kill_chain_phases
)
# print the created attack pattern object
print(attack_pattern)
```

```python
from stix2 import Relationship

# Create a relationship between the Campaign and attack pattern
relationship = Relationship(source_ref=campaign.id, relationship_type="uses", target_ref=attack_pattern.id)
print(campaign.serialize())

# print the relationship indicator & malware
relationship = Relationship(source_ref=indicator.id, relationship_type="indicates",target_ref=malware.id)
print(relationship)

# print the relationship indicator & campain
relationship = Relationship(source_ref=indicator.id, relationship_type="indicates",target_ref=campaign.id)
print(relationship)

# print the relationship campaign & identity
relationship = Relationship(source_ref=campaign.id, relationship_type="created by",target_ref=identity.id)
print(relationship)

# print the relationship campaign & malware
relationship = Relationship(source_ref=campaign.id, relationship_type="uses",target_ref=malware.id)
print(relationship)

# print the relationship campaign & vulnerability
relationship = Relationship(source_ref=campaign.id, relationship_type="targets",target_ref=vulnerability.id)
print(relationship)

# print the relationship campaign & attack pattern
relationship = Relationship(source_ref=campaign.id, relationship_type="uses",target_ref=attack_pattern.id)
print(relationship)

# print the relationship campaign & intrusion set
relationship = Relationship(source_ref=campaign.id, relationship_type="is attributed to",target_ref=intrusion_set.id)
print(relationship)

# print the relationship campaign & threat actor
relationship = Relationship(source_ref=campaign.id, relationship_type="is attributed to",target_ref=threat_actor.id)
print(relationship)
```
